## Decision Tree for Classification

Độ đo Gini có tể được xác định bởi công thức:

$$
Gini(D) = \frac{n_1}{n}Gini(D_1)+\frac{n_2}{n}Gini(D_2)
$$

$$
Gini(D_i)= 1-\displaystyle \sum_{j=1}^c p_j^2
$$

Đô đo Entropy được xác định bởi công thức:

$$
Entropy(D) =\frac{n_1}{n}Entropy(D_1) +\frac{n_2}{n}Entropy(D_2)
$$

$$
Entropy(D_i) = - \displaystyle \sum_{j=1}^2 p_jlog_2p_j
$$


Trong đó D là tập dữ liệu ban đầu có n phần tử và có thể được phân chia thành 2 tập con là $D_1$ có $n_1$ phần tử và $D_2$ có $n_2$ phần tử. $p_j$ là xác suất của cá sample trong $D_j$ thuộc vào class c.

Độ đo Information Gain được xác định bởi công thức:

$$
Gain(D) =1- Entropy(D)
$$


Để xác định được thuộc tính quan trọng với cây quyết định chúng ta có thể chọn Entropy hoặc Gini thấp nhất, Gini cao nhất .

a. Số dữ liệu trong hai lớp bằng nhau

$Gini(D) = -\frac{1}{2}*log_2(\frac{1}{2}) - \frac{1}{2}*log_2(\frac{1}{2}) $

$Gini(D) = -\frac{1}{2} *(-1) -\frac{1}{2}*(-1) =\frac{1}{2} +\frac{1}{2} = 1$

Câu hỏi 8 Sắp xếp các đoạn code phân loại trên bộ dữ liệu Iris dựa vào Decision Tree:

Thứ tự đúng là

a) C - B - A - D

b) A - B - D - C

c) D - C - A - B

d) A - C - D - B

Đáp án là a) C - B - A - D


```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

# Paragraph C :
# Load the diabetes dataset
iris_X, iris_y =  datasets.load_iris(return_X_y = True)
# Split train : test = 8:2
X_train, X_test, y_train, y_test = train_test_split(
	iris_X, iris_y,
	test_size = 0.2,
	random_state = 42)

# Paragraph B :
# Define model
dt_classifier = DecisionTreeClassifier()

# Paragraph A :
# Train
dt_classifier.fit(X_train, y_train)

# Paragraph D :
# Preidct and evaluate
y_pred = dt_classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```
